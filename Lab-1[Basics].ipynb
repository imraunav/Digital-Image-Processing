```python
import matplotlib.pyplot as plt
```

```python
path = "./misc/4.1.03.tiff"
im = plt.imread(path)
```

```python
plt.imshow(im)
plt.show()
```

```python
im.max(), im.min()
```

- It is sometimes needed to normalize the image values in the range[0, 1].
- For that, simply divide all elements by max possible value. For 8-bit image, its 255.

```python
im.shape
```

```python
r = im[:, :, 0]
g = im[:, :, 1]
b = im[:, :, 2]

plt.subplot(1, 3, 1)
plt.imshow(r, cmap='Oranges')
# plt.colorbar()

plt.subplot(1, 3, 2)
plt.imshow(g, cmap='Greens')
# plt.colorbar()

plt.subplot(1, 3, 3)
plt.imshow(b, cmap='Blues')
# plt.colorbar()

plt.tight_layout()
plt.show()
```

```python
grey = 0.2125*r + 0.7154*g + 0.07121*b
plt.imshow(grey, cmap="grey")
plt.show()
```

```python
import numpy as np
```

```python
f = np.zeros((40, 80, 3), dtype=np.uint8) # black rgb image 
```

```python
# red
m = 0*20
n = 3*20 
f[m:m+20, n:n+20] = [255, 0, 0]

# green
m = 1*20
n = 0*20 
f[m:m+20, n:n+20] = [0, 255, 0]

# blue
m = 1*20
n = 2*20 
f[m:m+20, n:n+20] = [0, 0, 255]

# white
m = 0*20
n = 1*20 
f[m:m+20, n:n+20] = [255, 255, 255]

# yellow
m = 0*20
n = 0*20 
f[m:m+20, n:n+20] = [255, 255, 0]

# magenta
m = 0*20
n = 2*20 
f[m:m+20, n:n+20] = [255, 0, 255]

# cyan
m = 1*20
n = 1*20 
f[m:m+20, n:n+20] = [0, 255, 255]


plt.imshow(f)
```
